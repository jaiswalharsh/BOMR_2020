```python
import cv2
import matplotlib.pyplot as plt
```

```python
filename = 'input_picture.png'
img = cv2.imread(filename, cv2.IMREAD_COLOR)
plt.imshow(img[:,:,::-1])
```

```python
canny_img = cv2.Canny(img, 60, 120)
plt.imshow(canny_img, cmap='gray');
plt.title('Canny'); plt.axis('off');
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors

filename = 'input_picture.png'
img = cv2.imread(filename, cv2.IMREAD_COLOR)
resized = cv2.resize(img, (50,50), interpolation = cv2.INTER_AREA) 
plt.figure(figsize=(30,50))
plt.subplot(121),plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(resized)
plt.title('Resized Image'), plt.xticks([]), plt.yticks([])

plt.show
```
